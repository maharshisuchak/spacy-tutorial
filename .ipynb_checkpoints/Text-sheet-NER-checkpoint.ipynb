### Named Entity Recognition

List out the names of the entities in a sheet.

```python
import csv
import spacy 
nlp=spacy.load("en")

```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("24sepcompar.csv")
dataset.columns=["1","2","3","4"]
textList=dataset["3"]
```

```python
for idx,text in enumerate(textList):
    if type(text)==str:
        tokens=nlp(text)
        total_list=[]
        ent_list1={} 
        ent_list2={}
        for token in tokens.ents:
            if token.label_ in ["GPE","ORG","PERSON"]:
                if ent_list1.get(token.label_)!=None:
                    ent_list1[token.label_]=ent_list1.get(token.label_)+", "+token.orth_.strip()
                else:
                     ent_list1[token.label_]=token.orth_.strip()
            else:
                if ent_list2.get(token.label_)!=None:
                    ent_list2[token.label_]=ent_list2.get(token.label_)+", "+token.orth_.strip()
                else:
                     ent_list2[token.label_]=token.orth_.strip()

            for key1 in ent_list1.keys():
                total_list.append(key1+": "+ent_list1.get(key1))


            total_list.append("\n")
            for key2 in ent_list2.keys():
                total_list.append(key2+": "+ent_list2.get(key2))
        dataset["Entities"][idx]=" ".join(total_list)
        print(idx,"Done")

print("Complete Task !!!")
dataset.to_csv("24sepcompar_with_entities.csv")
    
    
```
