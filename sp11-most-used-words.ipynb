### How to find most common words using Spacy and Python

```python
import spacy
from collections import Counter
```

```python
nlp = spacy.load('en')
```

```python
docx = nlp(open('/Users/happythoughts/Documents/tale-of-two-cities.txt').read())
```

```python
docx
```

```python
# Remove Punct,Stop
# Nouns
nouns = [ token.text for token in docx if token.is_stop != True and token.is_punct != True and token.text != "Chapter" and token.pos_ == 'NOUN' ]
```

```python
nouns
```

```python
word_freq = Counter(nouns)
```

```python
common_nouns = word_freq.most_common(100)
```

```python
print(common_nouns)
```

### Most common Verbs

```python
# Remove Punct,Stop
# Verbs
verbs = [ token.text for token in docx if token.is_stop != True and token.is_punct != True and token.pos_ == 'VERB' ]
```

```python
word_freq = Counter(verbs)
```

```python
common_verbs = word_freq.most_common(100)
```

```python
print(common_verbs)
```

```python
# Remove Punct,Stop
# Verbs
words = [ token.text for token in docx if token.is_stop != True and token.is_punct != True ]
```

```python
word_freq = Counter(words)
```

```python
common_words = word_freq.most_common(500)
```

```python
print(common_words)
```
